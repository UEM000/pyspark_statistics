```python
from pyspark.sql import SparkSession
from experiment import Experiment
import pandas as pd
```

```python
def initialization(path: str, session: SparkSession):
    df = (
            session.read.format('csv')
            .option("header", "true")
            .option("inferSchema", "true")
            .option("ignoreLeadingWhiteSpace", "true") 
            .option("ignoreTrailingWhiteSpace", "true")
            .load(path)
        )
    
    df = df.drop("_c0")

    return df

def session_init() -> SparkSession:
    session = (
            SparkSession.builder
            .master("local[*]")
            .config("spark.driver.memory", "4g")
            .config("spark.executor.memory", "4g")
            .config("spark.memory.fraction", "0.8") 
            .config("spark.memory.storageFraction", "0.3")
            .config("spark.jars.packages", "ch.cern.sparkmeasure:spark-measure_2.12:0.23") 
            .getOrCreate()
          )
    return session
session = session_init()
```

```python
# path = "../../data.csv"
path = "../../nyc_taxi_2020-01.csv"
result_path = "../test_results/"
output_path = "./arhive/test.txt"


data = initialization(path, session)

# data = pd.read_csv(path, index_col=0)

experiment = Experiment(session=session, data=data, result_path=result_path, report=True, output=False, output_path=output_path)
experiment.execute()

# session.stop()
```
