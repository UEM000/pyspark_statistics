```python
def mean_over_split(df: pd.DataFrame,
                    group_column: str,
                    target_columns: List[str],
                    k_split: int):
    stats = (
                df
                .loc[:, [group_column] + target_columns]
                .groupby(df[group_column].map(lambda x: x[k_split]))
                [target_columns]
                # .mean()
                .agg(["mean", "count", "std"])
                # .values
            )
    # return stats

    stats.index = pd.MultiIndex.from_product([[k_split], stats.index], names=['split', 'group'])
    return stats
    
# result = (
#             pd.DataFrame({"k_splits" : range(k_splits)})
#             .apply(lambda x: mean_over_split(testign, "group", "pre_spends", x["k_splits"]), axis=1)
#         )

# result
# .value_counts()
#                             .unstack()
#                             .fillna(0)
result = pd.concat([mean_over_split(testign, "group", ["pre_spends", "post_spends"], split_idx) for split_idx in range(k_splits)])
result
```

```python
def freq_over_splits(df: pd.DataFrame,
                     group_column: str,
                     target_columns: List[str],
                     k_split: int):
    stats = (
                    df
                    .loc[:, [group_column] + target_columns]
                    .groupby(df[group_column].map(lambda x: x[k_split]))
                    [target_columns]
                    .apply(lambda x: x.apply(pd.Series.value_counts))
                    .unstack()
                    .dropna(axis=1)

                    # .fillna(0)
                    # .assign(freq=lambda df_: sum([df_[column] for column in target_columns]))
                    # .loc[:, "freq"]
                    # .reset_index(1)
                    # .stack()
                    # .reset_index()
                    # .rename(columns=["label", "freq"])
                    # .unstack()
                    # .fillna(0)
                )

    stats.index = pd.MultiIndex.from_product([[k_split], stats.index], names=['split', 'group'])
    return stats
# freq_over_splits(b_new_df, "group", ["gender", "industry"], 0)
result = pd.concat([freq_over_splits(b_new_df, "group", ["gender", "industry"], split_idx) for split_idx in range(k_splits)])
result
```
